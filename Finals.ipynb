#  Creating the game

***Creating the class***

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
from collections import defaultdict
import time
import random

class Akinator:
    def __init__(self, path, node=0):
        self.path = path
        self.df = pd.read_csv(self.path)
        self.node = node
        self.mobs = None
        self.X = None
        self.Y = None
        self.tree = None
        self.ch_info = defaultdict(list)
        self.ans_count = 0
        self.random_s = random.randint(1, 20)
    def getting_f_t(self):
        self.mobs = list(self.df['Name'])
        name_to_number = {name: i for i, name in enumerate(self.mobs)}
        self.df['Name'] = self.df['Name'].map(name_to_number)
        self.Y = self.df['Name']
        self.X = self.df.drop('Name', axis=1)
        object_cols = self.X.select_dtypes(include='object').columns
        for col in object_cols:
            dummy_columns = self.X[col].str.get_dummies(', ')
            dummy_columns.columns = [f'{col}_{cold.strip()}' for cold in dummy_columns.columns]
            self.X = pd.concat([self.X, dummy_columns], axis=1)
            self.X.drop(col, axis=1, inplace=True)
        return self.mobs, self.X, self.Y
    def fit(self):
        clf = DecisionTreeClassifier(random_state=self.random_s)
        clf.fit(self.X, self.Y)
        self.tree = clf.tree_
        return self.tree
    def filter(self):
        self.df = pd.read_csv(self.path)
        for key, values in self.ch_info.items():
            included_values = [value for value in values if not value.startswith('Not ')]
            excluded_values = [value[4:] for value in values if value.startswith('Not ')]
        
            in_df = pd.DataFrame()
            ex_df = pd.DataFrame()
        
            if len(included_values) != 0:
                for value in included_values:
                    in_df = pd.concat([in_df, self.df[self.df[key].str.contains(value)]])
            else:
                pass
        
            if len(excluded_values) != 0:
                for value in excluded_values:
                    ex_df = pd.concat([ex_df, self.df[self.df[key].str.contains(value)]])
            else:
                pass
        
            self.df = pd.concat([self.df, ex_df, ex_df]).drop_duplicates(keep=False)
        
            self.df = pd.concat([in_df, self.df]).drop_duplicates()
        return self.df
    def game(self):
        if self.tree.children_left[self.node] == self.tree.children_right[self.node]:
            predicted_class = self.tree.value[self.node].argmax()
            while True:
                ans_f = input(f"\nI guess it's: {self.mobs[predicted_class]}\nAm I right? \n1. Yes\n2. No\n").lower()
                if ans_f == 'yes' or ans_f == '1':
                    print('Great, guessed right one more time!')
                    break
                else:
                    self.ans_count += 1
                    if self.ans_count == 2:
                        new = input('\nIt was a pleasure to play with you.\nWho are you thinking about?\nPlease provide me your character name: ')
                        self.ch_info['Name'].append(new)
                        break
                    else:
                        ans_con = input('\nWould you like to continue? \n1. Yes\n2. No\n').lower()
                        if ans_con == 'yes' or ans_con == '1':
                            self.ch_info = defaultdict(list)
                            self.df = pd.read_csv(self.path)
                            self.mobs, self.X, self.Y = self.getting_f_t()
                            self.node = 0
                            self.random_s += random.randint(1, 22)
                            self.tree = self.fit()
                            self.game()
                            break
                        else:
                            new = input('\nIt was a pleasure to play with you.\nWho are you thinking about?\nPlease provide me your character name: ')
                            self.ch_info['Name'].append(new)
                            break
            return 
    
        feature_index = self.tree.feature[self.node]
        feature_name = self.X.columns[feature_index]
        
        ans = input(f"\nIs the {feature_name.split('_')[0]} of you character is {feature_name.split('_')[1]}?\n1. Yes\n2. No\n3. Don't know\n4. Probably yes\n5. Probably not\n").lower()
        if ans == 'yes' or ans == '1':
            self.ch_info[f"{feature_name.split('_')[0]}"].append(feature_name.split('_')[1])
            self.node = self.tree.children_right[self.node]
            self.game()
        elif ans == 'no' or ans == '2':
            self.ch_info[f"{feature_name.split('_')[0]}"].append(f"Not {feature_name.split('_')[1]}")
            self.node = self.tree.children_left[self.node]
            self.game()
        elif ans == "don't know" or ans == '3':
            if len(self.ch_info) != 0:
                self.df = self.filter()
                self.mobs, self.X, self.Y = self.getting_f_t()
                self.random_s += random.randint(1, 22)
                self.tree = self.fit()
                self.game()
            else:
                self.random_s += random.randint(1, 22)
                self.tree = self.fit()
                self.game()
        elif ans == 'probably yes' or ans == '4':
            self.ch_info[f"{feature_name.split('_')[0]}"].append({feature_name.split('_')[1]})
            self.node = self.tree.children_right[self.node]
            self.game()
        elif ans == 'probably not' or ans == '5':
            self.ch_info[f"{feature_name.split('_')[0]}"].append(f"Not {feature_name.split('_')[1]}")
            self.node = self.tree.children_left[self.node]
            self.game()
        else:
            print("Please answer with above options.")
            self.game()
    def play(self):
        ans = input("Hello! Let's play a game. I will try to guess what you're thinking of. Are you ready? \n1. Yes\n2. No\n").lower()
        if ans == 'yes' or ans == '1':
            print('Here the some guides for questions:\n1. When question is asking about occuptaion, it refers to the ways of earning money by character in the past or present.\n2. Rank means same thing, i.e the rank the character has ever achieved in the Tv show episodes.')
            time.sleep(10)
            self.game()
        else:
            print("That's okay. Maybe next time! It's a pity you didn't want to play.")
    def get_info(self):
        return dict(self.ch_info)
```
